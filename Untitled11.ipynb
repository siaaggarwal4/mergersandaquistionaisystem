```python
from transformers import BertForSequenceClassification, AutoTokenizer, TrainingArguments, Trainer, DataCollatorWithPadding
from datasets import Dataset
```

```python
data = Dataset.from_csv("filtered copy.csv")
```

```python
model = BertForSequenceClassification.from_pretrained('nlpaueb/legal-bert-base-uncased', num_labels=12)
tokenizer = AutoTokenizer.from_pretrained('nlpaueb/legal-bert-base-uncased')
```

```python
unique= sorted(set(data["Type"]))
unique_type = {unique[i]: i for i in range(len(unique))}
unique_type
# unique_type2 = {uniqu: i for i, uniqu in enumerate(unique_type)}
# unique_type2
```

```python
nums = [unique_type[i["Type"]] for i in data]
print(nums)
data = data.add_column("label", nums)
data[6]
```

```python
data.remove_columns("Type")
```

```python
def tokenize(text):
  input_ids = tokenizer(text["Contract"], padding=True, truncation=True, max_length=256)
  input_ids["label"]= text["label"]
  return input_ids
```

```python
textch = [tokenize(text) for text in data]
input_ids = [i["input_ids"] for i in textch]
attn_mask = [i["attention_mask"] for i in textch]
# label = [i["label"] for i i]
data = data.add_column("input_ids", input_ids)
data = data.add_column("attention_mask", attn_mask)
# data.add_column("Tokenizedata",textch)
```

```python
dataseteval = Dataset.from_csv("filterdata2.csv")

dataseteval = dataseteval.add_column("label", [unique_type[i[" Type"]] for i in dataseteval])
tech = [tokenize(text) for text in dataseteval]

dataseteval = dataseteval.add_column("input_ids", [i['input_ids'] for i in tech])
dataseteval = dataseteval.add_column("attention_mask", [i["attention_mask"] for i in tech])
dataseteval = dataseteval.remove_columns("Unnamed: 0")
dataseteval = dataseteval.remove_columns("Contract")
dataseteval = dataseteval.remove_columns(" Type")
dataseteval
```

```python
trainer_ = TrainingArguments(output_dir="./model", num_train_epochs=2, per_device_train_batch_size=4, logging_dir="./logs")
datac = DataCollatorWithPadding(tokenizer=tokenizer)
trainer = Trainer(model=model, args=trainer_, data_collator=datac,  train_dataset=data, eval_dataset=dataseteval)
trainer.train()
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import numpy as np
def compute_metrics(eval_pred):
  logits, labels = eval_pred.predictions, eval_pred.label_ids
  preds = np.argmax(logits, axis=1)
  precision, recal, f1score, _ = precision_recall_fscore_support(labels, preds, average='binary')
  accuracy = accuracy_score(labels, preds)
  return {'f1_score': f1score, 'precision':precision, 'accuracy':accuracy, 'recal':recal}
```

```python
trainer.compute_metrics = compute_metrics
```

```python
trainer.evaluate(metric_key_prefix="eval")# dataset_test
```
